# Probability density function

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# Create time values (in seconds)
t = np.linspace(0, 10, 1000)

# Gamma distribution (shape=2, scale=1.5)
# Peaks around t=1, then decays
pdf_gamma = gamma.pdf(t, a=2, scale=1.5)

# Alternative: Lognormal distribution
# pdf_lognormal = lognormal.pdf(t, s=0.8, scale=1.5)

# Plot
plt.figure(figsize=(12, 6))

plt.plot(t, pdf_gamma, linewidth=2, color='red')
plt.xlabel('Time to Detection (seconds)', fontsize=11)
plt.ylabel('Probability Density', fontsize=11)
plt.title('Gamma Distribution (shape=2)', fontsize=12)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
