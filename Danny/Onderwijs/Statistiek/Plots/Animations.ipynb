```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import pandas as pd

# Customize plot styles in a uniform manner
plt.style.use('plot_stylesheet.mplstyle') 
```

```python
# Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create figure
fig, ax = plt.subplots()
line1, = ax.plot([], [], color='blue', label='Sine Wave', linewidth=2)
line2, = ax.plot([], [], color='red', linestyle='--', linewidth=2, label='Cosine Wave')

# Initialize function
def init():
    ax.set_xlim(0, 10)
    ax.set_ylim(-1.5, 1.5)
    return line1, line2

# Animation function
def update(frame):
    if frame < len(x):
        line1.set_data(x[:frame], y1[:frame])
    if frame >= len(x):
        line2.set_data(x[:frame-len(x)], y2[:frame-len(x)])
    return line1, line2

# Create animation
ani = FuncAnimation(fig, update, frames=np.arange(0, 2*len(x)), init_func=init, blit=True)

# Save animation as a gif
ani.save('overlay_animation.gif', writer='imagemagick')

plt.show()
```
