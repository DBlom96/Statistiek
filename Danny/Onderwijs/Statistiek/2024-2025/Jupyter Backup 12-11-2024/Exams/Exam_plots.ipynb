```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm, expon, t, chi2, uniform, binom, poisson
from datetime import datetime

plt.rcParams.update(plt.rcParamsDefault)
# plt.style.use('../Styles/stylesheet.mplstyle') 
```

```python
def generate_arbitrary_continuous_dist(low=0, high=10, func=lambda x: norm(x, 0, 1), xlabel=r"$x$", ylabel=r"$f(x)$", title=r"Kansdichtheidsfunctie $f(x)$"):
    x_dist = np.linspace(low, high, 10_000)
    y_dist = func(x_dist)
    
    print(x_dist, y_dist)
    
    plt.plot(x_dist, y_dist)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    
    timestamp = datetime.now().strftime("%Y%m%d-%H%M%S")
    plt.savefig(f"Exam-{timestamp}.pdf")
```

```python
def dist1(arr):       
    return np.array(list(map(lambda a: 1.0/3888 * a**3 * (36 - a**2) * (0 <= a <= 6), arr)))

generate_arbitrary_continuous_dist(0, 6, dist1)
```
